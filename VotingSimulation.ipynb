```python
#Votação

windowsServer = []
unix = []
linux = []
netware = []
macOS = []
outro = []
valido = []
invalido = []
total = []

voto = 1000

while voto != 0:
    voto = int(input("Digite seu voto: "))
    if voto == 1:
        windowsServer.append(voto)
        total.append(voto)
        valido.append(voto)
    elif voto == 2:
        unix.append(voto)
        total.append(voto)
        valido.append(voto)
    elif voto == 3:
        linux.append(voto)
        total.append(voto)
        valido.append(voto)
    elif voto == 4:
        netware.append(voto)
        total.append(voto)
        valido.append(voto)
    elif voto == 5:
        macOS.append(voto)
        total.append(voto)
        valido.append(voto)
    elif voto == 6:
        outro.append(voto)
        total.append(voto)
        valido.append(voto)
    elif voto == 0:
        print("Eleição encerrada!")
    else:
        invalido.append(voto)
        total.append(voto)

win100 = 0
win100 = len(windowsServer*100)/len(valido)
unix100 = 0
unix100 = len(unix*100)/len(valido)
linux100 = 0
linux100 = len(linux*100)/len(valido)
net100 = 0
net100 = len(netware*100)/len(valido)
mac100 = 0
mac100 = len(macOS*100)/len(valido)
outro100 = 0
outro100 = len(outro*100)/len(valido)
total00 = 0
total00 = (win100 + unix100 + linux100 + net100 + mac100 + outro100)

resultado = [len(windowsServer), len(unix), len(linux), len(netware), len(macOS), len(outro)]

#anunciar resultado fazendo um if manipulando max(resultado) == candidato por exemplo

if max(resultado) == len(windowsServer):
    print("\nVencedor: Windows Server")
    print("Quantidade de votos: {} votos.".format(len(windowsServer)))
    print("Porcentagem: {}% dos votos válidos.".format(win100))
elif max(resultado) == len(unix):
    print("\nVencedor: Unix")
    print("Quantidade de votos: {} votos.".format(len(unix)))
    print("Porcentagem: {}% dos votos válidos.".format(unix100))
elif max(resultado) == len(linux):
    print("\nVencedor: Linux")
    print("Quantidade de votos: {} votos.".format(len(linux)))
    print("Porcentagem: {}% dos votos válidos.".format(linux100))
elif max(resultado) == len(netware):
    print("\nVencedor: Netware")
    print("Quantidade de votos: {} votos.".format(len(netware)))
    print("Porcentagem: {}% dos votos válidos.".format(net100))
elif max(resultado) == len(macOS):
    print("\nVencedor: Mac OS")
    print("Quantidade de votos: {} votos.".format(len(macOS)))
    print("Porcentagem: {}% dos votos válidos.".format(mac100))
else:
   print("\nOutro sistema operacional desconhecido venceu a votação!")

#Bolar uma solução pra anunciar empate tecnico

print("\nSistema Operacional  votos    %")
print("-------------------  -----  ------")
print("1 - Windows Server         {}     {}% ".format(len(windowsServer), win100))
print("2 - Unix                   {}     {}% ".format(len(unix), unix100))
print("3 - Linux                  {}     {}% ".format(len(linux), linux100))
print("4 - Netware                {}     {}% ".format(len(netware), net100))
print("5 - Mac OS                 {}     {}% ".format(len(macOS), mac100))
print("6 - Outro                  {}     {}% ".format(len(outro), outro100))
print("-------------------  -----  ------")
print("Total                  {}     {}% ".format(len(total), total00))
print("\nTotal de eleitores: {}".format(len(total)))
print("Total de votos nulos: {}".format(len(invalido)))
print("Total de votos válidos: {} ".format(len(valido)))

```
